1 Gray level transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Lena.jpg", cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# 1. Negative Transformation
negative_image = 255 - image

# 2. Contrast Stretching
min_val, max_val = np.min(image), np.max(image)
contrast_stretched_image = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)

# 3. Thresholding
_, thresholded_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# 4. Gamma Correction
gamma = 2.0  # Change this value to experiment with different gamma levels
gamma_corrected_image = np.array(255 * (image / 255) ** gamma, dtype='uint8')

# Display the original and transformed images
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(2, 3, 2)
plt.title("Negative Transformation")
plt.imshow(negative_image, cmap='gray')

plt.subplot(2, 3, 3)
plt.title("Contrast Stretching")
plt.imshow(contrast_stretched_image, cmap='gray')

plt.subplot(2, 3, 4)
plt.title("Thresholding")
plt.imshow(thresholded_image, cmap='gray')

plt.subplot(2, 3, 5)
plt.title("Gamma Correction")
plt.imshow(gamma_corrected_image, cmap='gray')

plt.tight_layout()
plt.show()
```

3 Linear Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Lena.jpg", cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# Parameters for linear transformation
alpha = 1.2  # Contrast control (1.0-3.0)
beta = 30    # Brightness control (0-100)

# Apply the linear transformation
linear_transformed_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Display the original and transformed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Linear Transformed Image")
plt.imshow(linear_transformed_image, cmap='gray')

plt.tight_layout()
plt.show()
```

4 Foggy image and Gamma Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image (color image for foggy effect)
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Lena.jpg")

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# ---- Foggy Image Effect ----
fog_intensity = 0.5  # Adjust the intensity of the fog
# Create a white image of the same size as the original
white_image = np.full(image.shape, 255, dtype=np.uint8)
# Blend the original image with the white image to create fog effect
foggy_image = cv2.addWeighted(image, 1 - fog_intensity, white_image, fog_intensity, 0)

# ---- Power-Law (Gamma) Transformation ----
# Convert to grayscale for gamma correction
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gamma = 2.0  # Change this value to experiment with different gamma levels
gamma_corrected_image = np.array(255 * (gray_image / 255) ** gamma, dtype='uint8')

# Display the original, foggy, and gamma-corrected images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title("Foggy Image")
plt.imshow(cv2.cvtColor(foggy_image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 3)
plt.title("Gamma Corrected Image")
plt.imshow(gamma_corrected_image, cmap='gray')

plt.tight_layout()
plt.show()
```

5 Histogram Equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Lena.jpg", cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and equalized images
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(2, 2, 2)
plt.title("Equalized Image")
plt.imshow(equalized_image, cmap='gray')

# Plot the histograms
plt.subplot(2, 2, 3)
plt.title("Histogram of Original Image")
plt.hist(image.ravel(), 256, [0, 256])
plt.xlim([0, 256])

plt.subplot(2, 2, 4)
plt.title("Histogram of Equalized Image")
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

6 Calculate and display the histogram 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Lena.jpg", cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# Calculate the histogram
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Display the image and its histogram
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Grayscale Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Histogram")
plt.plot(histogram)
plt.xlim([0, 256])
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

7 Implement the image enhancement techniques

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Lena.jpg", cv2.IMREAD_COLOR)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Histogram Equalization
equalized_image = cv2.equalizeHist(gray_image)

# Contrast Adjustment (using alpha and beta values)
alpha = 1.5  # Contrast control (1.0-3.0)
beta = 0     # Brightness control (0-100)
contrast_adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Sharpening the image using kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharpened_image = cv2.filter2D(image, -1, kernel)

# Displaying the images
titles = ['Original Image', 'Gray Image', 'Histogram Equalized', 'Contrast Adjusted', 'Sharpened Image']
images = [image, gray_image, equalized_image, contrast_adjusted, sharpened_image]

plt.figure(figsize=(10, 10))
for i in range(5):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray' if i == 1 or i == 2 else 'viridis')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()

# Save the results
cv2.imwrite('gray_image.jpg', gray_image)
cv2.imwrite('equalized_image.jpg', equalized_image)
cv2.imwrite('contrast_adjusted.jpg', contrast_adjusted)
cv2.imwrite('sharpened_image.jpg', sharpened_image)
```
