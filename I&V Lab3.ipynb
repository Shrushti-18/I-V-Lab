1 Create this g1(x,y)signal and visualize it in 2-D & 3-D viwes. Compute the 2D DFT of g1(x,y) and display the 2D and 3D plot
 of magnitude spectrum of G1(u,y)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the signal g1(x, y) - Gaussian function for example
def g1(x, y):
    return np.exp(-(x**2 + y**2))

# Create a grid of x, y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute g1(x, y)
Z = g1(X, Y)

# 2D plot of g1(x, y)
plt.figure()
plt.title("2D view of g1(x, y)")
plt.imshow(Z, extent=[-5, 5, -5, 5], cmap='viridis')
plt.colorbar()
plt.show()

# 3D plot of g1(x, y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D view of g1(x, y)")
plt.show()

# Compute the 2D DFT of g1(x, y)
G1 = np.fft.fftshift(np.fft.fft2(Z))
magnitude_spectrum = np.abs(G1)

# 2D plot of the magnitude spectrum of G1(u, v)
plt.figure()
plt.title("2D Magnitude Spectrum of G1(u, v)")
plt.imshow(np.log(1 + magnitude_spectrum), extent=[-5, 5, -5, 5], cmap='viridis')
plt.colorbar()
plt.show()

# 3D plot of the magnitude spectrum of G1(u, v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.log(1 + magnitude_spectrum), cmap='viridis')
plt.title("3D Magnitude Spectrum of G1(u, v)")
plt.show()
```

 2 Create this g2(x,y) signal and visualize it in 2-D and 3-D views. Compute the 2D DFT of g2(x,y) and display the 2D and 3D      plot of magnitude spectrum of G2(u,v)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the signal g2(x, y) - Combination of sinusoidal and Gaussian function
def g2(x, y):
    return np.sin(2 * np.pi * x) * np.exp(-(x**2 + y**2))

# Create a grid of x, y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute g2(x, y)
Z = g2(X, Y)

# 2D plot of g2(x, y)
plt.figure()
plt.title("2D view of g2(x, y)")
plt.imshow(Z, extent=[-5, 5, -5, 5], cmap='viridis')
plt.colorbar()
plt.show()

# 3D plot of g2(x, y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D view of g2(x, y)")
plt.show()

# Compute the 2D DFT of g2(x, y)
G2 = np.fft.fftshift(np.fft.fft2(Z))
magnitude_spectrum = np.abs(G2)

# 2D plot of the magnitude spectrum of G2(u, v)
plt.figure()
plt.title("2D Magnitude Spectrum of G2(u, v)")
plt.imshow(np.log(1 + magnitude_spectrum), extent=[-5, 5, -5, 5], cmap='viridis')
plt.colorbar()
plt.show()

# 3D plot of the magnitude spectrum of G2(u, v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.log(1 + magnitude_spectrum), cmap='viridis')
plt.title("3D Magnitude Spectrum of G2(u, v)")
plt.show()
```

4 Compute 2D DCT of an image and Reconstruct the image using the inverse DCT function.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.fftpack import dct, idct

# Function to perform 2D DCT
def dct2(a):
    return dct(dct(a.T, norm='ortho').T, norm='ortho')

# Function to perform 2D inverse DCT
def idct2(a):
    return idct(idct(a.T, norm='ortho').T, norm='ortho')

# Load the grayscale image
image = cv2.imread(r"C:\Users\Shrushti S G\Downloads\Tiger.jpeg", cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded properly
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# Compute the 2D DCT of the image
dct_image = dct2(image)

# Display the DCT of the image
plt.figure()
plt.title("2D DCT of the image")
plt.imshow(np.log(np.abs(dct_image) + 1), cmap='gray')
plt.colorbar()
plt.show()

# Reconstruct the image using inverse DCT
reconstructed_image = idct2(dct_image)

# Display the reconstructed image
plt.figure()
plt.title("Reconstructed image using inverse DCT")
plt.imshow(np.clip(reconstructed_image, 0, 255), cmap='gray')
plt.colorbar()
plt.show()
```
